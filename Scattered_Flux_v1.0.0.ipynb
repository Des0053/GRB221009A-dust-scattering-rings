**Definitions of functions**

```python
def lorentz(x,a,b,c):
    return a*1./(np.pi*c)*(c**2/((x-b)**2.+c**2))

def ring_rad(t,d_layer_norm,t0, d_source, z):
    return 8.3*(d_layer_norm/10**(-6.))**(-0.5)*(d_source/100.)**(-1./2.)*((t-t0)*(3600.*24.)/(10**4.))**(1./2.)*(1.+z)**(-0.5) #arcmin

def time_delay(theta, d_layer_norm, d_source, z):
    return 1e4*(theta/8.3)**2*(d_source/100.)*(d_layer_norm/1e-6)*(1+z)

def d_dist(theta,dt, d_source, z):
    return (8.3/theta)**2*(d_source/100.)**(-1)*(dt*(3600.*24.)/10**4)*(1+z)**(-1)*10**(-6)

def theta_sc(E_keV,a):
    return 1.04/(E_keV*a) # arcmin

def n(a,q,no):
    return no*a**(-q)

def ntot(amin,amax,no,q):
    if q != 1:
        ntot = no*(amax**(-q+1)-amin**(-q+1))/(-q+1)
    return 

def S(E,Epk,ph_index,So):
    return So*(E/Epk)**(-ph_index+1)

def fint(a,q,E,x,theta):
    return a**(6-q)*np.exp(-theta**2/(2.*(1.-x)**2*(theta_sc(E,a))**2))
```
